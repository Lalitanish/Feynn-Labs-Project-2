```python
import plotly.express as px

# Assuming your DataFrame is named 'target_age'
data_25_to_34 = target_age[(target_age['Age'] >= 25) & (target_age['Age'] <= 34)]
data_35_to_45 = target_age[(target_age['Age'] >= 35) & (target_age['Age'] <= 45)]

labels = ['Age 25-34', 'Age 35-45']
sizes = [len(data_25_to_34), len(data_35_to_45)]
colors = ['pink', 'salmon']

fig = px.pie(names=labels, values=sizes, color_discrete_sequence=colors,
             title="Percentage of People in Age Groups 25-34 and 35-45",
             hole=0.7)
fig.show()

```

```python
# Calculate the percentage of each fuel type
fuel_percentage = new_dataframe2['fuel'].value_counts(normalize=True) * 100

# Light colors for visualization
colors = {
    'Petrol': 'lightpurple',
    'Diesel': 'lightblue',
    'CNG': 'lightsalmon',
    'LPG': 'lightgreen',
    'Electric': 'lightpink'
}

# Plotting the percentage of each fuel type with light colors
fig = px.pie(names=fuel_percentage.index, values=fuel_percentage.values,
             title='Percentage of Fuel Types in Car Listings',
             labels={'names': 'Fuel Type', 'values': 'Percentage'},
             color_discrete_map=colors,
             hole=0.7)
fig.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler


# Selecting features and target variable
X = target_age[['Salary', 'Wife Salary', 'Total Salary']]
y = target_age['Price']

# Standardize the features using StandardScaler
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Split the standardized data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_standardized, y, test_size=0.3, random_state=35)

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the standardized test set
y_pred = model.predict(X_test)

# Create an interactive scatter plot with plotly
fig = px.scatter(x=y_test, y=y_pred, labels={'x': 'Actual Price', 'y': 'Predicted Price'},
                 title='Actual vs Predicted Prices (Standardized)', template='plotly_dark')

# Add a diagonal line for reference
fig.add_shape(
    type='line',
    x0=min(y_test), y0=min(y_test),
    x1=max(y_test), y1=max(y_test),
    line=dict(color='red', dash='dash')
)

# Show the plot
fig.show()
```
